# TWITTER SCRAPER

1) Importar librerías

```python
import twint
import pandas as pd
import numpy as np
```

2) Consulta para obtener todos los tweets que incluyen los tèrminos coronavirus y costa rica desde el 01 de febrero. Guardar el resultado de forma local

```python
c = twint.Config()
c.Search = "coronavirus AND \"costa rica\""
c.Retweets = True
c.Store_json = True
c.User_full = True
c.Store_csv = True
c.Output = "Tweets_Coronavirus.csv"
c.Since = "2020-02-01 00:00:00"
c.Hide_output = True
twint.run.Search(c)
```

3) Leer MisTweets.csv con un nuevo id

```python
datos = pd.read_csv('Tweets_Coronavirus.csv')
datos["tweet_id"] = datos.index + 1
datos.head(10)
```

4) Tomar la columna original del dataframe que tiene la lista de usuarios mencionados en el tweet y crear un nuevo dataset que contiene por cada fila a un usuario mencionado junto con el identificador del tweet respectivo

```python
tweets_usuarios=pd.DataFrame(columns=["tweet_id","username"])
usuario=datos[["tweet_id","mentions"]]
for n1 in usuario.iterrows():           
    n3=n1[1][1].replace("[", "").replace("]", "").replace("'", "").replace(",", "").split()
    for n4 in n3:
        tweets_usuarios=tweets_usuarios.append({"tweet_id" : n1[1][0] , "username" : n4} , ignore_index=True)
```

5) Tomar la columna de usuarios que escriben el tweet y fundirlos con la lista generada en 4). Luego se eliminan los repetidos.

```python
usuario=datos["username"].values.tolist()+tweets_usuarios["username"].values.tolist()
usuario=list(set(usuario))
len(usuario)
```

6) Generar el archivo con los nodos de usuarios únicos (USUARIO.CSV)

```python
usuario=pd.DataFrame(data=usuario,columns=["username"])
usuario["usuario_id"] = usuario.index + 1
usuario.to_csv("usuarios.csv",index=False)
usuario.head(10)
```

7) Agregar al dataframe original el usuario_id del dataframe con la lista de usuarios únicos. Generar el archivo con las relaciones de tweets que MENCIONAN usuarios (MENCIONA.CSV)

```python
tweets_usuarios=tweets_usuarios.merge(usuario, on='username')
tweets_usuarios[["tweet_id","usuario_id"]].to_csv("menciona.csv",index=False)
tweets_usuarios.head(10)
```

8) Tomar la columna original del dataframe que tiene la lista de hashtags contenidos en el tweet y crear un nuevo dataset que contiene por cada fila a un hashtag junto con el identificador del tweet respectivo

```python
tweets_hashtags=pd.DataFrame(columns=["tweet_id","hashtags"])
hashtag=datos[["tweet_id","hashtags"]]
for n1 in hashtag.iterrows():           
    n3=n1[1][1].replace("[", "").replace("]", "").replace("'", "").replace(",", "").split()
    for n4 in n3:
        tweets_hashtags=tweets_hashtags.append({"tweet_id" : n1[1][0] , "hashtags" : n4} , ignore_index=True)
```

9) Generar el archivo con los nodos de hashtags únicos (HASHTAGS.CSV)

```python
hashtags=list(set(tweets_hashtags["hashtags"].tolist()))
hashtags=pd.DataFrame(data=hashtags,columns=["hashtags"])
hashtags["hashtag_id"] = hashtags.index + 1
hashtags.to_csv("hashtags.csv",index=False)
hashtags.head(10)
```

10) Agregar al dataframe tweets_hashtags originado en 8) el hashtag_id del dataframe con la lista de hashtags únicos. Generar el archivo con las relaciones de tweets que CONTIENEN hashtags (CONTIENE.CSV)

```python
tweets_hashtags=tweets_hashtags.merge(hashtags, on='hashtags')
tweets_hashtags[["tweet_id","hashtag_id"]].to_csv("contiene.csv",index=False)
tweets_hashtags.head(10)
```

11) Tomar la columna original del dataframe que tiene la lista de urls contenidos en el tweet y crear un nuevo dataset que contiene por cada fila a un url junto con el identificador del tweet respectivo (Se extrae solamente el dominio).

```python
tweets_urls=pd.DataFrame(columns=["tweet_id","urls"])
url=datos[["tweet_id","urls"]]
for n1 in url.iterrows():           
    n3=n1[1][1].replace("[", "").replace("]", "").replace("'", "").replace(",", "").split()
    for n4 in n3:
        n5=n4.replace("http://", "").replace("https://", "").split("/")[0]
        print(n5)
        tweets_urls=tweets_urls.append({"tweet_id" : n1[1][0] , "urls" : n5} , ignore_index=True)
```

12) Generar el archivo con los nodos de enlaces únicos (SITIOS.CSV)

```python
urls=list(set(tweets_urls["urls"].tolist()))
urls=pd.DataFrame(data=urls,columns=["urls"])
urls["url_id"] = urls.index + 1
urls.to_csv("sitios.csv",index=False)
urls
```

13) Agregar al dataframe tweets_urls originado en 11) el url_id del dataframe con la lista de urls únicos. Generar el archivo con las relaciones de tweets que REFERENCIAN urls (REFERENCIA.CSV)

```python
tweets_urls=tweets_urls.merge(urls, on='urls', how="left")
tweets_urls[["tweet_id","url_id"]].to_csv("referencia.csv",index=False)
tweets_urls
```

14) Generar el archivo con los nodos de tweets (TWEETS.CSV)

```python
datos[["tweet_id","tweet"]].replace({"\\\\":""}, regex=True).to_csv("tweets.csv",index=False)
```

15) Generar el archivo con las relaciones de usuarios que ESCRIBEN tweets (ESCRIBE.CSV)

```python
datos=datos.merge(usuario, on='username')
datos[["tweet_id","usuario_id"]].to_csv("escribe.csv",index=False)
```
